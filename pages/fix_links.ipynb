```python
import os


def fix_links(line: str) -> str:
    if r"[[" in line:
        start_link = line.index('[')
        end_link = line.index(']')
        link = line[start_link+2:end_link].split("|")
        try:
            md_link = f"[{link[1]}]({link[0]})"
        except IndexError:
            md_link = str(link)
        line = line[:start_link] + md_link + line[end_link+2:]
        return line 
    else:
        return line


def fix_file(file_name: str) -> None:    
    fixed_text = ""
    with open(file_name, "r") as original_file:
        for line in original_file:
            # Clean up the line
            line = line.strip()

            # Fix headers
            # ------------------------------------
            # Turn ** headers into ### headers
            if line.startswith("**"):
                line = line[2:-2]
                line = "## " + line
            
            for i in range(1, 7):
                check_head = ("=" * i + " ")
                if line.startswith(check_head):
                    print(line[i:-1*i])
                
            # Replace dokuwiki links with plain markdown links
            line = fix_links(line)
            
            # Add back     
            fixed_text += line + "\n"
    
    print(fixed_text)
    # os.rename(file_name, file_name + ".backup")
    # with open(file_name, "w") as fixed_file:
    #     fixed_file.write(fixed_text)


print(fix_file("./lang/python.md"))
```
